**변경고민/시도하지않은 사항**
- fl_gamma 변수: it's better not to modify this param because increasing this parameter can push precision down and recall up drastically.
  - reference thread: https://github.com/ultralytics/yolov5/issues/553

## 전체 데이터 학습 17&exp17

전처리 수정
- hyp.scratch-low.yaml에서 변경
  - X-ray 이미지로 Hue나 Saturation 전처리는 적절치 않으나, ```hsv_v``` value(brightness) 전처리는 적절.
      - X-ray 이미지는 그레이 스케일이므로 적용을 조심히 할 필요성이 있음. Brightness shift를 0.4 -> 0.1 로 줄임
      - 0.1과 같은 작은 범위는 다양한 수준의 밝기와 대비를 시뮬레이션하는 데 도움이 될 수 있지만 너무 많은 변형은 비현실적인 훈련 샘플로 이어질 수 있습니다.
  - ```translate```매개변수는 이미지를 수평 또는 수직으로 이동할 수 있는 이미지 크기의 비율을 의미함.
    - 0.1이면 이미지가 너비 및 높이의 최대 10%+/-까지 변환 될 수 있음을 의미함.
    - 때문에 0.05라는 작은 숫자로 변경
  -  ```scale```매개변수는 줌인 줌아웃을 의미하며, 0.5는 +/-50% 비율로 줌인 줌아웃을 의미한다. 즉, 50% 혹은 150%가 될 수 있음.
    - 해부학적 구조를 왜곡할 수 있으므로, 의료이미지에는 적절치 않음
    - 보수적인 스케일링 범위 안에서 줌인 줌아웃 진행이 적절하므로, 0.5에서 0.2로 변경
  - ```mosaic```매개변수는 여러 이미지를 하나의 이미지로 만들어주는 기능.
    - 작은 물체가 많아져 YOLO의 단점을 극복하도록 도와주지만, BOX 일부가 잘릴경우 객체의  인식혼동이슈가 있을 수 있음.
    - 따라서 1.0 -> 0.0으로 변경
  -  ```fliplr``` 1.0에서 0.0으로 변경
